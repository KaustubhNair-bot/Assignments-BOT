```python
import pandas as pd
import numpy as np
```

```python
import json
```

```python
#reading a json file::
with open("demo_json.json","r") as f:
    data=json.load(f)
print(f"json data is loaded,length of the data is : {len(data)}")
```

```python
data
```

```python
#flattening a json format data into a dataframe::
df = pd.json_normalize(
    data,
    record_path=["items"],          
    meta=[
        "order_id",
        "order_date",
        ["customer", "customer_id"],
        ["customer", "name"],
        ["customer", "city"],
        ["payment", "method"],
        ["payment", "status"],
    ]
)
```

```python
df
```

```python
#cleaning a data::
```

```python
column_mapping = {
    "product_id": "Product ID",
    "product_name": "Product Name",
    "category": "Product Category",
    "price": "Price",
    "quantity": "Quantity Sold",
    "order_id": "Order ID",
    "order_date": "Order Date",
    "customer.customer_id": "Customer ID",
    "customer.name": "Customer Name",
    "customer.city": "Customer City",
    "payment.method": "Payment Method",
    "payment.status": "Order Status",
}
df1 = df.rename(columns=column_mapping)
df1
```

```python
#orderDate coloumn should be in datetime format::
df1["Order Date"] = pd.to_datetime(df1["Order Date"])
```

```python
df1
```

```python
# price & quantity sold column should be in numeric::
df1["Price"] = pd.to_numeric(df1["Price"])
df1["Quantity Sold"] = pd.to_numeric(df1["Quantity Sold"])
```

```python
df1
```

```python
# Handle missing values::
df1["Customer City"] = df1["Customer City"].fillna("Unknown")
df1["Customer Name"] = df1["Customer Name"].fillna("Unknown Customer")
df1["Price"] = df1["Price"].fillna(0)
df1["Quantity Sold"] = df1["Quantity Sold"].fillna(0)
```

```python
df1
```

```python
print("Cleaned DataFrame info:")
df.info()
```

```python
#numpy based transformation::
```

```python
price_arr = df1["Price"].to_numpy()
qty_arr   = df1["Quantity Sold"].to_numpy()
status_arr = df1["Order Status"].to_numpy()
```

```python
print(f"the price_arr is : {price_arr}")
print(f"the qty_arr is : {qty_arr}")
print(f"the status_arr is : {status_arr}")
```

```python
#Adding total amount coloumn::
total_amount_arr = price_arr * qty_arr
df1["Total Amount"] = total_amount_arr
```

```python
df1
```

```python
#adding Revenue coloumn::
revenue_arr = np.where(status_arr == "Success", total_amount_arr, 0)
df1["Revenue Recognized"] = revenue_arr
```

```python
df1
```

```python
df_success = df1[df1["Order Status"] == "Success"].copy()
order_revenue = df_success.groupby("Order ID")["Revenue Recognized"].sum()
print(df_success)
print(order_revenue)
```

```python
# Identify high vhalue order IDs
high_value_order_ids = order_revenue[order_revenue > 1000].index

order_id_arr = df1["Order ID"].to_numpy()
is_high_value_arr = np.isin(order_id_arr, high_value_order_ids)

df1["High Value Order Flag"] = np.where(is_high_value_arr, "Yes", "No")
```

```python

print("Data with NumPy-based calculated fields:")
df1[[
    "Order ID", "Product Name", "Price", "Quantity Sold",
    "Total Amount", "Order Status", "Revenue Recognized",
    "High Value Order Flag"
]]
```

```python
df_success = df1[df1["Order Status"] == "Success"].copy()

df1["Order Completion Flag"] = np.where(df1["Order Status"] == "Success", "Completed", "Not Completed")

print("Sample of business logic columns:")
df1[[
    "Order ID", "Order Status", "Order Completion Flag",
    "Total Amount", "Revenue Recognized", "High Value Order Flag"
]]
```

```python
mandatory_columns = [
    "Order ID",
    "Order Date",
    "Customer ID",
    "Customer Name",
    "Customer City",
    "Product Name",
    "Product Category",
    "Quantity Sold",
    "Total Amount",
    "Payment Method",
    "Order Status",
    "High Value Order Flag",
]

final_df = df1[mandatory_columns].copy()

final_df = final_df.sort_values(by=["Order Date", "Order ID"]).reset_index(drop=True)

print("Final analytical DataFrame:")
final_df
```

```python
#Business Questions to Answer::
```

```python
df_success = df1[df1["Order Status"] == "Success"].copy()

print("Business Questions & Answers")

# 1> Total revenue per city::
revenue_per_city = df_success.groupby("Customer City")["Revenue Recognized"].sum().sort_values(ascending=False)
print("\n1. Total revenue per city:")
print(revenue_per_city)
```

```python
# 2> Product category with highest revenue::
revenue_per_category = df_success.groupby("Product Category")["Revenue Recognized"].sum()
top_category = revenue_per_category.idxmax()
print("\n2. Product category with highest revenue:")
print(revenue_per_category)
print(f"Highest revenue category: {top_category}")
```

```python
# 3> Count of high-value orders ::
high_value_order_ids = df1[df1["High Value Order Flag"] == "Yes"]["Order ID"].unique()
high_value_order_count = len(high_value_order_ids)
print(f"\n3. Count of high-value orders: {high_value_order_count}")
```

```python
# 4> Average order value on successful orders only::
order_level_revenue = df_success.groupby("Order ID")["Revenue Recognized"].sum()
total_revenue = order_level_revenue.sum()
num_orders = order_level_revenue.shape[0]
aov = total_revenue / num_orders if num_orders > 0 else 0
print(f"\n4. Average order value (AOV): {aov:.2f}")
```

```python
# 5> Most preferred payment method::
payment_counts = df1.groupby("Payment Method")["Order ID"].nunique()
top_payment_method = payment_counts.idxmax()
print("\n5. Most preferred payment method (unique orders):")
print(payment_counts)
print(f"Most preferred payment method: {top_payment_method}")
```

```python
# 6> Top customers by revenue::
customer_revenue = df_success.groupby(["Customer ID", "Customer Name"])["Revenue Recognized"].sum()
top_customers = customer_revenue.sort_values(ascending=False).head(5)
print("\n6. Top customers by revenue:")
print(top_customers)
```

```python
# 7> Completed vs cancelled/failed orders::
order_status_counts = df1.groupby("Order Status")["Order ID"].nunique()
print("\n7. Completed vs cancelled/failed orders (unique orders):")
print(order_status_counts)
```
